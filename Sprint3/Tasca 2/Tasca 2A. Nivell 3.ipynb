### Nivell 3: Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#Exercici 8 
img=mpimg.imread('apple.png')

#Accedint al canal R, eliminem el vermell
no_red = np.copy(img)
no_red[:, :, 0]=0
plt.show(plt.imshow(no_red))

#Accedint al canal G, eliminem el verd
no_green = np.copy(img)
no_green[:, :, 1]=0
plt.show(plt.imshow(no_green))

#Accedint al canal B, eliminem el blau 
no_blue = np.copy(img)
no_blue[:, :, 2]=0
plt.show(plt.imshow(no_blue))

#Mitjançant mpimg.save() guardem les imatges modificades
mpimg.imsave("nored_apple.png", no_red)
mpimg.imsave("nogreen_apple.png", no_green)
mpimg.imsave("noblue_apple.png", no_blue)

```
