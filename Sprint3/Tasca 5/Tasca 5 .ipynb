```python
!pip install pandas
import pandas as pd
import numpy as np 
```

```python
dly_flights = pd.read_csv('DelayedFlights.csv')
#Amb .shape podem saber quantes files i columnes conté el data set
dly_flights.shape

#Eliminem la columna Unnamed, ja que no ens aporta cap dada rellevant
temp_df = dly_flights.drop(['Unnamed: 0'], axis=1)

#També podem obtenir quants valors nuls presenta cada columna i eliminem aquelles columnes que en presenten molts
temp_df.isnull().sum()
temp_df = temp_df.drop(['CarrierDelay','WeatherDelay', 'NASDelay','SecurityDelay', 'LateAircraftDelay'], axis=1)



```

```python
#Finalment ens quedem amb aquelles columnes més rellevants per nosaltres i per el que ens demana l'exercici:
final_df = temp_df[['Year', 'Month', 'DayofMonth','UniqueCarrier', 'FlightNum', 'AirTime', 'ArrDelay', 'DepDelay', 'Origin', 'Dest', 'Distance' ]]
```

```python
#A continuació eliminem aquelles files que presenten dades nul·les per alguna columna
final_df = final_df.dropna()
final_df.isnull().sum()
```

```python
#CREACIÓ DE NOVES COLUMNES
#Primer extreiem en variables les columnes que utilitzarem per crear-ne de noves, passem milles a km i min a h
air_time = final_df['AirTime']/60
distance = final_df['Distance']*1.61
#Afegim la columna de velocitat mitja
final_df['Speed(Km/h)'] = distance/air_time

#Per determinar si un vol arriba tard o no definim la següent funció i l'apliquem a la columna ArrDelay
#Es crearà una nova columna amb 0(no delay) i 1(delay)
def tard(x):
    if x <= 0:
        return 0
    else:
        return 1
final_df['ArrDelay(Y/N)'] = final_df['ArrDelay'].apply(tard)

final_df.head(10)
```

```python
#AEROLíNIES AMB MÉS ENDARRERIMENTS:
#Aquí he considerat com endarreriment aquells vols que arriben amb retràs independentment de la seva sortida:
#Generem diferents grups segons el codi(UniqueCarrier) de l'aerolínia
airlines = final_df.groupby('UniqueCarrier')

#A continuació iterem per els diferents grups generats i declarem 2 llistes que contindrà, 1 el codi de l'aerolínia
#i l'altra el total de vols endarrerits per cada aerolínea
tot_delay = []
airline_code = []
for airline, group in airlines:
    delay = sum(group['ArrDelay(Y/N)'])
    tot_delay.append(delay)
    airline_code.append(airline)

#Amb les dades obtingudes anteriorment, generem un nou data frame i l'ordenem per el nº d'endarreriments     
df = pd.DataFrame({'Airline': airline_code, 'TotDelyFlights': tot_delay})
df = df.sort_values('TotDelyFlights',ascending=False)
df
```

```python
#VOLS MÉS LLARGS:
#Creem una nova columna que conté l'origen i el destí
final_df['Route'] = final_df['Origin'] + '-' + final_df['Dest']
#Generem diferents grups segons les diferents rutes existents i seguim la metodologia de l'exercici anterior
routes = final_df.groupby('Route')

ori_dest = []
distances = []
for route, group in routes:
    ori_dest.append(route)
    distances.append(group['Distance'].agg(np.mean))

distances_df = pd.DataFrame({'Route': ori_dest, 'Distance': distances})
distances_df = distances_df.sort_values('Distance',ascending=False)
distances_df.head(10)
    
```

```python
#VOLS MÉS ENDARRERITS:
ori_dest = []
delays = []
for route, group in routes:
    ori_dest.append(route)
    delays.append(sum(group['ArrDelay(Y/N)']))

delays_df = pd.DataFrame({'Route': ori_dest, 'TotDelyFlights': delays})
delays_df = delays_df.sort_values('TotDelyFlights',ascending=False)
delays_df.head(10)
```
